```python
import numpy as np
import pandas as pd

import librosa
import librosa.display
import soundfile as sf

import sklearn
from sklearn.model_selection import train_test_split

import IPython.display as ipd

import os
import pathlib
from PIL import Image

from sklearn.preprocessing import LabelEncoder, StandardScaler
import keras
from keras import layers
from keras import layers
import keras
from keras.models import Sequential


%matplotlib inline
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

##Loading Audio files

```python
audio_data = "gruesome.mp3"
X, sr = librosa.load(audio_data)
print(type(X), type(sr))
print(X.shape, sr)
```

```python
# play the audio
ipd.Audio(audio_data)
```

```python
# plot the audio array
plt.figure(figsize=(14, 5))
librosa.display.waveplot(X, sr=sr)
```

##Spectrogram

```python
# display a spectrogram
X = librosa.stft(X)  # converts data into short term Fourier transform
xdb = librosa.amplitude_to_db(abs(X))

plt.figure(figsize=(14, 5))
librosa.display.specshow(xdb, sr=sr, x_axis="time", y_axis="hz")  # display spectrogram
plt.colorbar()
```

```python
# convert the frequency axis to a logarithmic one
plt.figure(figsize=(14, 5))
librosa.display.specshow(xdb, sr=sr, x_axis="time", y_axis="log")  # display spectrogram
plt.colorbar()
```

```python
# let's create an Audio Signal
sr = 22050   # sample rate
T = 5.0      # seconds
t = np.linspace(0, T, int(T * sr), endpoint=False)  # time variable
x = 0.5 * np.sin(2 * np.pi * 220 * t)               # pure sine wave at 220 Hz

# Playing the audio
ipd.Audio(x, rate=sr)
```

```python
# Saving the audio
sf.write("tone_220.wav", x, sr)
```
